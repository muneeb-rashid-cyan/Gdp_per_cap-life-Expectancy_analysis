**Population with years**

```python
f, ax = plt.subplots(1, 5, figsize=(20, 4))
sns.set_style('darkgrid')
pop_per_cont = gapminder.groupby(['continent', 'year']).sum()['pop'].unstack(level=0)
pop_per_cont

#Population over time, Plots and Labels
for i, x in enumerate(pop_per_cont.columns):
    ax[i].plot(pop_per_cont[x])
    ax[i].set_xlabel('Years')
    ax[0].set_ylabel('Population')
    ax[i].set_title(str(x))


plt.show()
```

**Life Expactancy with years**

```python
life_per_cont = gapminder.groupby(['continent', 'year']).mean()['lifeExp'].unstack(level=0)


f, ax = plt.subplots(1, 5, figsize=(20, 4), sharey=True)
sns.set_style('darkgrid')


#Life Expectancy over time, Plots and Labels
for i, x in enumerate(life_per_cont.columns):
    ax[i].plot(life_per_cont[x])
    ax[i].set_xlabel('Years')
    ax[0].set_ylabel('Life Expectancy (years)')
    ax[i].set_title(str(x))


plt.show()
```
